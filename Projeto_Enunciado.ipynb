## APS2: **Classificação**

> Deadline: 15 Maio 2026

**Previsão de Renda Anual**

Desenvolver e avaliar modelos de classificação binária para prever se um indivíduo ganha mais de $50K$ por ano (`income > 50K`) ou não, com base nas características demográficas, educacionais e ocupacionais presentes no Adult Census Income Dataset.

Esta tarefa possui alta relevância prática em áreas como políticas públicas, recrutamento, análise de desigualdade social e estudos socioeconômicos.

As features com maior poder discriminativo esperado são: `education-num`, `capital-gain`, `capital-loss`, `hours-per-week`, `age`, `marital-status`, `occupation` e `relationship`. O aluno deve analisar explicitamente quais variáveis melhor separam as duas classes.


- Métricas de avaliação obrigatórias (todas devem ser reportadas):

    - Accuracy;
    - Precision, Recall e F1-score (macro, weighted e por classe);
    - ROC-AUC (área sob a curva ROC);
    - PR-AUC (área sob a curva Precision-Recall) – especialmente importante devido ao desbalanceamento;
    - Confusion Matrix com análise detalhada de falsos positivos e falsos negativos.

Elementos do projeto:

1. Técnicas de balanceamento: class weights, SMOTE, undersampling ou threshold moving (com justificativa);

1. Feature Engineering criativo (obrigatório criar ao menos 3 novas features). Exemplos sugeridos:

    - `capital_net` = `capital-gain` - `capital-loss`
    - `work_experience_proxy` = `age` - `education-num` - $6$;
    - Interações: `education-num` × `occupation`, `marital-status` × `age`, `sex` × `hours-per-week`


1. Modelagem:

    - Mínimo 3 modelos distintos, incluindo:

        - Dummy;
        - Baseline: Linear Regression / Ridge / Lasso;
        - Ensemble: Random Forest Regressor, XGBoost Regressor;

    - Validação cruzada robusta (Repeated K-Fold $5×$ ou $10×$ com random_state fixo).

1. Otimização e seleção de modelo:

    - Hyperparameter tuning (GridSearchCV ou Optuna/RandomizedSearchCV);
    - Comparação estatística entre modelos.

1. Análises de métricas sobre o modelo.

1. Interpretação e explicabilidade.

1. Discussão crítica obrigatória:

    - Limitações do modelo (multimodalidade da target, heterocedasticidade, causalidade vs. correlação);
    - Sugestões de melhoria e próximos passos.


Instruções adicionais:

- O código deve ser bem organizado, com comentários explicativos para cada etapa;
- As visualizações devem ser claras, com títulos, rótulos e legendas adequados;
- O relatório deve ser conciso, focando nos insights mais relevantes e nas decisões de pré-processamento;
- O grupo deve justificar todas as escolhas de parâmetros;
- O relatório deve incluir uma seção de conclusões, resumindo os principais insights e as estratégias propostas para a modelagem, além de discutir as limitações do dataset e os desafios que podem surgir durante o uso do modelo;
- Python obrigatório, utilizando bibliotecas como Pandas, NumPy, Matplotlib, Seaborn, Scikit-learn, entre outras para análise e visualização dos dados.
